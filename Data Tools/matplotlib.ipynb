## sub plots

```python
co=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]
d = [10,11,12,13,15,17,19]

plt.plot(co,a)
plt.plot(co,b)
plt.plot(co,c)
plt.plot(co,d)

plt.show()
```

```python
co=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]
d = [10,11,12,13,15,17,19]

plt.subplot(2,2,1)
plt.plot(co,a)
plt.subplot(2,2,2)
plt.plot(co,b)
plt.subplot(2,2,3)
plt.plot(co,c)
plt.subplot(2,2,4)
plt.plot(co,d)

plt.show()
```

```python
co=[1,2,3,4,5,6,7]
a=[50,51,52,48,47,49,46]
b=[43,42,40,44,33,35,37]
c=[45,48,48,46,40,42,41]
d = [10,11,12,13,15,17,19]

plt.subplot(3,2,1)
plt.plot(co,a)
plt.subplot(3,2,2)
plt.plot(co,b)
plt.subplot(3,2,3)
plt.plot(co,c)
plt.subplot(3,2,4)
plt.plot(co,d)

plt.show()
```

## pie chart

```python
x = [40,30,20,10,1]

y = ['Java','Python','c#','PHP','c++']

plt.pie(x,labels=y)
plt.show()
```

```python
x=[33,33,33]
y=['vishwa','srihitha','kumar']

plt.pie(x,labels=y)
plt.show()
```

```python
plt.figure(figsize=(10,8))
x=[33,33,33]
y=['vishwa','srihitha','kumar']

plt.pie(x,labels=y,shadow=True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
x = [40,30,20,10,1]

y = ['Java','Python','c#','PHP','c++']


plt.pie(x,labels=y,autopct = '%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,8))
x=[33,33,33]
y=['vishwa','srihitha','kumar']

plt.pie(x,labels=y,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,8))
x=[33,33,33]
y=['vishwa','srihitha','kumar']

plt.pie(x,labels=y,autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
x=[33,33,33]
y=['vishwa','srihitha','kumar']

plt.pie(x,labels=y,autopct='%1.1f%%',shadow=True,explode=[0,0,0.4])
plt.show()
```
