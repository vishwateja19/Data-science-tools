```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
df
```

## joint_plot(bivariate analysis=using 2 columns)

```python
sns.jointplot(x='total_bill',y='tip',data=df)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='g')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='g',kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='g',kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='g',kind='hist')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='g',kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,color='g',kind='resid')
```
